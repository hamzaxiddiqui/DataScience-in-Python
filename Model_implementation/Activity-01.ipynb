## Implementation of Linear ,KNN & Desicion tree regression for Boston House Price

```python
#import libraries and data set
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsRegressor 
from sklearn import tree
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
boston=load_boston()
predictors=boston.data
outcome=boston.target
```

```python
print('Features of Boston House:',np.shape(predictors))
print('Target of Boston House:',np.shape(outcome))
```

## Linear Regression

```python
LR=LinearRegression()
kfold=KFold(3,True,1)
k=1
LR_testing_avg_mse=[]
for train,test in kfold.split(predictors):
    train_mse=[]
    test_mse=[]
    LR.fit(predictors[train],outcome[train])
    
    train_set_predict=LR.predict(predictors[train])
    train_mse.insert(len(train_mse),mean_squared_error(outcome[train],train_set_predict))
    
    test_set_predict=LR.predict(predictors[test])
    test_mse.insert(len(test_mse),mean_squared_error(outcome[test],test_set_predict))
    
    LR_testing_avg_mse.append(sum(test_mse)/len(test_mse))
    print('\nFor %s-Fold validation' %(k))
    print('Train set Error',train_mse)
    print('Test set Error',test_mse)
    print('Predicted value for Test set',test_set_predict)
    k+=1
```

## KNN-Regressor

```python
kfold=KFold(3,True,1)
knn_r_testing_avg_mse=[]
for k in range(3,20,2):
    knn_r=KNeighborsRegressor(n_neighbors=k)
    train_mse=[]
    test_mse=[]
    for train,test in kfold.split(predictors):
       
        knn_r.fit(predictors[train],outcome[train])
        
        train_set_predict=knn_r.predict(predictors[train])
        
        train_mse.append(mean_squared_error(outcome[train],train_set_predict))
        
        test_set_predict=knn_r.predict(predictors[test])
        
        test_mse.append(mean_squared_error(outcome[test],test_set_predict))
    knn_r_testing_avg_mse.append(sum(test_mse)/len(test_mse))
    print('\nFor K = %s' %(k))   
    print('Train set Mean Squared Error =',train_mse)
    print('Test set Mean Squared Error =',test_mse)
    print('Predicted value for test set\n',test_set_predict)
    
        
```

## Decision Tree Regressor

```python
DT_R=tree.DecisionTreeRegressor()
DT_R_testing_avg_mse=[]
k=1
for train,test in kfold.split(predictors):
    train_mse=[]
    test_mse=[]
    
    DT_R.fit(predictors[train],outcome[train])
    
    train_set_predict=DT_R.predict(predictors[train])
    train_mse.append(mean_squared_error(train_set_predict,outcome[train]))
    
    test_set_predict=DT_R.predict(predictors[test])
    test_mse.append(mean_squared_error(test_set_predict,outcome[test]))
    
    DT_R_testing_avg_mse.append(sum(test_mse)/len(test_mse))
    print('\nFor %s-Fold'% (k))
    print('Train set Mean Square Error',train_mse)
    print('Test set Mean Square Error',test_mse)
    k+=1

```
